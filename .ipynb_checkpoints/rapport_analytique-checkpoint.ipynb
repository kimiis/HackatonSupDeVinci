## 1. Contexte et objectifs

Le changement climatique constitue l'un des défis majeurs du XXIe siècle. La France métropolitaine est particulièrement exposée : hausse des températures deux fois plus rapide que la moyenne mondiale, multiplication des événements extrêmes (canicules, inondations, sécheresses), montée du niveau de la mer sur ses 5 500 km de côtes.

**Objectif de cette étude :** construire un pipeline de données climatiques reproductible, entraîner des modèles prédictifs sur les indicateurs clés, et produire des projections à horizon 2030, 2050 et 2100 selon trois scénarios alignés sur les trajectoires du GIEC.

**Périmètre :** France métropolitaine, période 1900–2025 (historique), 2026–2100 (projections).

## 3. Analyse historique du territoire

````python
# ── 3.1 Température : tendance séculaire ──────────────────────────────────────
df_temp = df[['annee', 'temp_moy_france']].dropna()
df_temp = df_temp[df_temp['temp_moy_france'] > 9.5]  # filtre années incomplètes

baseline = df_temp[df_temp['annee'] <= 1920]['temp_moy_france'].mean()
moy10    = df_temp.set_index('annee')['temp_moy_france'].rolling(10, center=True).mean()

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(df_temp['annee'], df_temp['temp_moy_france'],
        color='#aaaaaa', lw=0.8, alpha=0.7, label='Mesure annuelle')
ax.plot(moy10.index, moy10.values,
        color='#fb923c', lw=2, label='Tendance (moy. mobile 10 ans)')
ax.axhline(baseline, color='#60a5fa', lw=1.5, ls='--',
           label=f'Niveau 1900-1920 ({baseline:.1f}°C)')

derniere_temp = df_temp['temp_moy_france'].iloc[-1]
ax.annotate(f'+{derniere_temp - baseline:.2f}°C\nde réchauffement',
            xy=(df_temp['annee'].iloc[-1], derniere_temp),
            xytext=(1990, baseline + 0.5),
            arrowprops=dict(arrowstyle='->', color='white'),
            color='white', fontsize=10)

ax.set_title('Température moyenne annuelle en France (1900–2025)', fontsize=13, pad=12)
ax.set_xlabel('Année'); ax.set_ylabel('°C')
ax.legend(); ax.grid(True)
plt.tight_layout()
plt.savefig('data/resultats/fig_temp_historique.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Hausse constatée : +{derniere_temp - baseline:.2f}°C depuis 1900-1920')
````

````python
# ── 3.2 Vue d'ensemble de tous les indicateurs ────────────────────────────────
indicateurs = {
    'CO₂ (ppm)':                'co2_ppm',
    'Niveau mer (mm)':           'niveau_mer_mm',
    'Jours chauds ≥30°C':        'jours_chauds_30',
    'Jours de gel ≤0°C':         'jours_gel',
    'Empreinte (t CO₂/hab)':     'empreinte_tCO2_hab',
    'Vendanges (DOY)':           'jour_vendanges',
    'Coût catastrophes (Mrd$)':  'dommages_Mrd_USD',
}
indicateurs = {k: v for k, v in indicateurs.items() if v in df.columns}

n = len(indicateurs)
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()
couleurs = ['#fb923c', '#60a5fa', '#f87171', '#4ade80', '#c084fc', '#a78bfa', '#fbbf24']

for i, (label, col) in enumerate(indicateurs.items()):
    ax = axes[i]
    sub = df[['annee', col]].dropna()
    ax.plot(sub['annee'], sub[col], color=couleurs[i], lw=1.5)
    ax.set_title(label, fontsize=10)
    ax.set_xlabel('Année', fontsize=8)
    ax.tick_params(labelsize=8)
    ax.grid(True, alpha=0.4)

# Masquer le dernier subplot si impair
for j in range(n, len(axes)):
    axes[j].set_visible(False)

fig.suptitle('Évolution des indicateurs climatiques — France', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('data/resultats/fig_indicateurs_overview.png', dpi=150, bbox_inches='tight')
plt.show()
````

````python
# ── 3.3 Corrélations entre indicateurs ───────────────────────────────────────
cols_corr = ['temp_moy_france', 'co2_ppm', 'niveau_mer_mm',
             'jours_chauds_30', 'empreinte_tCO2_hab', 'jour_vendanges']
cols_corr = [c for c in cols_corr if c in df.columns]

df_corr = df[cols_corr].dropna()
corr    = df_corr.corr()

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(corr, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(cols_corr))); ax.set_xticklabels(cols_corr, rotation=45, ha='right', fontsize=9)
ax.set_yticks(range(len(cols_corr))); ax.set_yticklabels(cols_corr, fontsize=9)

for i in range(len(cols_corr)):
    for j in range(len(cols_corr)):
        ax.text(j, i, f'{corr.iloc[i, j]:.2f}',
                ha='center', va='center', fontsize=8,
                color='black' if abs(corr.iloc[i, j]) < 0.6 else 'white')

ax.set_title('Matrice de corrélation des indicateurs climatiques', fontsize=12)
plt.tight_layout()
plt.savefig('data/resultats/fig_correlations.png', dpi=150, bbox_inches='tight')
plt.show()

print('Corrélations avec la température :')
print(corr['temp_moy_france'].drop('temp_moy_france').sort_values(ascending=False).to_string())
````

## 4. Méthodologie de modélisation

### 4.1 Choix du modèle : Meta Prophet

Prophet est un modèle de prévision de séries temporelles développé par Meta (Facebook) en 2017. Il est particulièrement adapté à notre problème pour plusieurs raisons :

| Critère | Raison du choix |
|---|---|
| Données annuelles longues (100+ ans) | Prophet gère nativement les longues séries avec changements de régime |
| Tendances non-linéaires | Détection automatique des *changepoints* (ruptures de pente) |
| Données manquantes | Robuste aux NaN — pas besoin d'interpolation préalable |
| Intervalles de confiance | Estimation bayésienne via Stan → incertitude croissante vers le futur |
| Interprétabilité | Décomposition explicite tendance + saisonnalité |

### 4.2 Équation du modèle

$$y(t) = g(t) + s(t) + h(t) + \epsilon_t$$

Avec :
- $g(t)$ : tendance (linéaire par morceaux avec changepoints)
- $s(t)$ : saisonnalité (désactivée — données annuelles)
- $h(t)$ : effets des événements particuliers (non utilisé)
- $\epsilon_t$ : bruit résiduel

### 4.3 Évaluation : cross-validation temporelle (walk-forward)

```
│←────── initial (60% de la série) ──────→│← horizon →│
│                  fold 1                  │  5 ans     │
│                       fold 2             │  5 ans     │
│                             fold 3       │  5 ans     │
─────────────────────────────────────────────────────────→ temps
1900                                                    2025
```

Métriques calculées à chaque fold, puis moyennées :
- **RMSE** = $\sqrt{\frac{1}{n}\sum(\hat{y}-y)^2}$ — sensible aux grandes erreurs
- **MAE** = $\frac{1}{n}\sum|\hat{y}-y|$ — robuste aux valeurs aberrantes  
- **MAPE** = $\frac{1}{n}\sum\frac{|\hat{y}-y|}{|y|}\times 100$ — erreur relative en %

### 4.4 Scénarios climatiques

| Scénario | Anomalie cible 2100 | Équivalent GIEC | Description |
|---|---|---|---|
| Optimiste | +1.4°C | SSP1-1.9 | Efforts très ambitieux, décarbonation rapide |
| Intermédiaire | +2.7°C | SSP2-4.5 | Politiques actuelles maintenues |
| Pessimiste | +4.4°C | SSP5-8.5 | Aucune action climatique (BAU) |

La trajectoire est une interpolation linéaire entre la dernière observation et la cible 2100.

## 5. Résultats et performances des modèles

````python
# ── 5.1 Tableau des performances ──────────────────────────────────────────────
import os

path_perf = 'data/resultats/comparaison_performances.csv'
if os.path.exists(path_perf):
    df_perf = pd.read_csv(path_perf)
    print('Performances des modèles Prophet par indicateur :')
    print(df_perf.to_string(index=False))
    display(df_perf.style
            .format({'rmse': '{:.4f}', 'mae': '{:.4f}', 'mape': '{:.2f}'})
            .background_gradient(subset=['rmse', 'mae'], cmap='RdYlGn_r')
            .background_gradient(subset=['mape'], cmap='RdYlGn_r')
            .set_caption('Plus la valeur est faible, meilleur est le modèle'))
else:
    print('Lancer modele_prophet.py pour générer les performances.')
````

````python
# ── 5.2 Visualisation forecast température ────────────────────────────────────
fc_path = 'data/resultats/forecast_temperature.csv'
if os.path.exists(fc_path):
    fc_temp = pd.read_csv(fc_path)
    fc_temp['annee'] = pd.to_numeric(fc_temp['annee'], errors='coerce')

    fig, ax = plt.subplots(figsize=(14, 5))

    # Observations
    df_t = df[['annee', 'temp_moy_france']].dropna()
    df_t = df_t[df_t['temp_moy_france'] > 9.5]
    ax.plot(df_t['annee'], df_t['temp_moy_france'],
            'white', lw=1.5, label='Observations réelles', zorder=5)

    # Forecast
    fc_fut = fc_temp[fc_temp['annee'] > df_t['annee'].max()]
    ax.fill_between(fc_fut['annee'], fc_fut['yhat_lower'], fc_fut['yhat_upper'],
                    alpha=0.2, color='steelblue', label='Intervalle de confiance 95%')
    ax.plot(fc_fut['annee'], fc_fut['yhat'],
            color='#60a5fa', lw=2, ls='--', label='Projection Prophet')

    # Scénarios
    sc_path = 'data/resultats/scenarios_temperature.csv'
    if os.path.exists(sc_path):
        df_sc = pd.read_csv(sc_path)
        annee_ref = int(df_t['annee'].iloc[-1])
        temp_ref  = df_t['temp_moy_france'].iloc[-1]
        for sc, coul, label in [
            ('optimiste',     '#4ade80', 'Optimiste +1.4°C'),
            ('intermediaire', '#fb923c', 'Intermédiaire +2.7°C'),
            ('pessimiste',    '#f87171', 'Pessimiste +4.4°C'),
        ]:
            rows = df_sc[df_sc['scenario'] == sc]
            xs = [annee_ref] + rows[rows['annee'] > annee_ref]['annee'].tolist()
            ys = [temp_ref]  + rows[rows['annee'] > annee_ref]['temp_proj_C'].tolist()
            ax.plot(xs, ys, color=coul, lw=2.5, marker='o', ms=7, label=label)

    ax.set_title('Température France — Observations et projections 2100', fontsize=13)
    ax.set_xlabel('Année'); ax.set_ylabel('°C')
    ax.legend(loc='upper left', fontsize=9); ax.grid(True)
    plt.tight_layout()
    plt.savefig('data/resultats/fig_forecast_scenarios.png', dpi=150, bbox_inches='tight')
    plt.show()
````

---
## 6. Projections 2030 / 2050 / 2100

````python
# ── 6.1 Tableau des projections par scénario ──────────────────────────────────
sc_path = 'data/resultats/scenarios_temperature.csv'
if os.path.exists(sc_path):
    df_sc = pd.read_csv(sc_path)
    pivot = df_sc.pivot(index='annee', columns='scenario',
                        values=['temp_proj_C', 'anomalie_C'])
    pivot.columns = [f'{b} — {a}' for a, b in pivot.columns]
    print('Projections de température (°C) :')
    display(pivot.round(2))
````

````python
# ── 6.2 Projections des autres indicateurs ────────────────────────────────────
resultats_proj = {}
for annee_cible in [2030, 2050, 2100]:
    row = {}
    for nom, label in [
        ('co2',               'CO₂ (ppm)'),
        ('niveau_mer',        'Niveau mer (mm)'),
        ('empreinte_carbone', 'Empreinte (t CO₂/hab)'),
        ('vendanges',         'Vendanges (DOY)'),
    ]:
        path = f'data/resultats/forecast_{nom}.csv'
        if os.path.exists(path):
            fc = pd.read_csv(path)
            val = fc[fc['annee'] == annee_cible]['yhat']
            row[label] = round(val.values[0], 1) if not val.empty else None
    resultats_proj[annee_cible] = row

df_proj = pd.DataFrame(resultats_proj).T
df_proj.index.name = 'Horizon'
print('Projections Prophet (tendance extrapolée) :')
display(df_proj)
````

## 7. Recommandations climat-territoire

### 7.1 Synthèse des risques identifiés

L'analyse des données historiques et des projections permet d'identifier quatre risques prioritaires pour la France métropolitaine :

| Risque | Signal observé | Horizon de criticité | Référence PNACC-3 |
|---|---|---|---|
| **Canicules et vagues de chaleur** | +2.6°C depuis 1900-1920, multiplication des jours TX≥30°C | Dès 2030 | Axe 2 — Santé |
| **Sécheresse et tension hydrique** | Vendanges avancées de plusieurs semaines, projections de hausse | 2030-2050 | Axe 1 — Eau |
| **Montée du niveau de la mer** | +120 mm moyens sur les ports français vs 1961-1990 | 2050-2100 | Axe 4 — Littoral |
| **Dérèglement des écosystèmes** | Indicateurs phénologiques (vendanges) en décalage croissant | En cours | Axe 5 — Biodiversité |

### 7.2 Recommandations par acteur

#### Pour les collectivités territoriales
- Intégrer les projections climatiques dans les PLU (plans locaux d'urbanisme)
- Développer les trames vertes et bleues pour réduire les îlots de chaleur
- Adapter les plans de prévention des risques naturels (PPRN) aux nouvelles projections
- Renforcer les dispositifs de vigilance canicule et sécheresse

#### Pour les citoyens
- Réduire l'empreinte carbone individuelle sous 5 t CO₂/hab (objectif intermédiaire 2030)
- Adapter le logement (isolation, végétalisation, récupération d'eau)
- Modifier les pratiques alimentaires (réduction viande rouge, circuits courts)
- Participer aux démarches collectives de végétalisation urbaine

#### Pour les entreprises et acteurs économiques
- Réaliser un bilan carbone et définir une trajectoire de décarbonation (SNBC)
- Anticiper les risques climatiques dans les plans de continuité d'activité
- Investir dans les infrastructures résilientes (eau, énergie)

### 7.3 Alignement avec les cadres réglementaires

| Cadre | Objectif | Lien avec notre analyse |
|---|---|---|
| PNACC-3 (2024) | Adapter la France au changement climatique | Indicateurs de risque calés sur les axes du PNACC |
| SNBC (2020) | Neutralité carbone en 2050 | Empreinte carbone individuelle tracée et projetée |
| Accord de Paris | Limiter le réchauffement à +1.5°C | Scénario optimiste aligné sur cet objectif |
| Earth Action Report 2025 | Actions concrètes sur eau, énergie, nature | Préconisations citoyennes du dashboard |

````python
# ── Consultation des runs MLflow ───────────────────────────────────────────────
import mlflow

try:
    client = mlflow.tracking.MlflowClient()
    exp    = client.get_experiment_by_name('hackathon26_climat_france')
    if exp:
        runs = client.search_runs(exp.experiment_id,
                                  order_by=['metrics.rmse ASC'])
        rows = []
        for r in runs:
            rows.append({
                'modèle':  r.data.tags.get('mlflow.runName', r.info.run_id),
                'RMSE':    r.data.metrics.get('rmse'),
                'MAE':     r.data.metrics.get('mae'),
                'MAPE %':  r.data.metrics.get('mape'),
                'n_obs':   r.data.params.get('n_observations'),
                'période': f"{r.data.params.get('annee_debut')}–{r.data.params.get('annee_fin')}"
            })
        df_mlflow = pd.DataFrame(rows)
        print('Runs MLflow enregistrées :')
        display(df_mlflow)
    else:
        print('Aucune expérience MLflow trouvée. Lancer modele_prophet.py.')
except Exception as e:
    print(f'MLflow non disponible : {e}')
````
